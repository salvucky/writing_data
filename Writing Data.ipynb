```python
import pandas as pd
%matplotlib inline
```

## Reading File
##### Reads data from writing.csv which contains number of journal pages written every day from 6/10/2018-6/9/2019

```python
dates = pd.read_csv("writing.csv")
dates.head()
```

## Filtering the Data
#### Filters dates based on whether or not the quantity of journal pages written is greater than 0

```python
page_query_mask = dates['Number of pages written'] > 0
valid_dates = dates[page_query_mask]
valid_dates
```

## Dividing Data by Month
#### Sorts data by month and outputs a description for each

```python
june_query_mask = valid_dates["Date"].str.startswith("6")
june = valid_dates[june_query_mask]

july_query_mask = valid_dates["Date"].str.startswith("7")
july = valid_dates[july_query_mask]

august_query_mask = valid_dates["Date"].str.startswith("8")
august = valid_dates[august_query_mask]

september_query_mask = valid_dates["Date"].str.startswith("9")
september = valid_dates[september_query_mask]

october_query_mask = valid_dates["Date"].str.startswith("10")
october = valid_dates[october_query_mask]

november_query_mask = valid_dates["Date"].str.startswith("11")
november = valid_dates[november_query_mask]

december_query_mask = valid_dates["Date"].str.startswith("12")
december = valid_dates[december_query_mask]

january_query_mask = valid_dates["Date"].str.startswith("1/")
january = valid_dates[january_query_mask]

february_query_mask = valid_dates["Date"].str.startswith("2")
february = valid_dates[february_query_mask]

march_query_mask = valid_dates["Date"].str.startswith("3")
march = valid_dates[march_query_mask]

april_query_mask = valid_dates["Date"].str.startswith("4")
april = valid_dates[april_query_mask]

may_query_mask = valid_dates["Date"].str.startswith("5")
may = valid_dates[may_query_mask]

print("January", january.describe())
print("February", february.describe())
print("March", march.describe())
print("April", april.describe())
print("May", may.describe())
print("June", june.describe())
print("July", july.describe())
print("August", august.describe())
print("September", september.describe())
print("October", october.describe())
print("November", november.describe())
print("December", december.describe())
```
